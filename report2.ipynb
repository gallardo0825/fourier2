```python
import math 
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
n = 359
t = np.linspace(0,2*np.pi,n+1)#tは0から2piまでを360等分
fn = np.zeros(n+1)
for i in range(n+1):
   if t[i] <= np.pi:
       fn[i] = -1+2*t[i]/np.pi
   else:
       fn[i] = 3-2*t[i]/np.pi
ck = np.zeros(n+1)
for k in range (n+1):
    for i in range(n+1):
        ck[k] += np.real(fn[i]*np.exp(-(1.j)*2*np.pi*k*i/n))/n
rcvrfn = np.zeros(n+1) #rcvrfnはckから復元するfn
for i in range(n+1):
   for k in range(n+1):
       rcvrfn[i] += np.real(ck[k]*np.exp(1.j*2*np.pi*k*i/n))
plt.plot(t,rcvrfn)
```
