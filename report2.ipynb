```python
import math 
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
n = 360
t = np.linspace(0,2*np.pi,n)#tは0から2piまでを360等分
fn = np.zeros(n)
for i in range(n):
   if t[i] <= np.pi:
       fn[i] = -1+2*t[i]/np.pi
   else:
       fn[i] = 3-2*t[i]/np.pi
nck = np.zeros(n)
ck = np.zeros(n)
for k in range (n):
    for i in range(n):
        nck[k] += np.real(fn[i]*np.exp(-(1.j)*2*np.pi*k*i/n))
    ck[k] = nck[k]/n
        
```
